# Axes Customizing

1. fig.tight_layout
2. fig.subplots_adjust
3. Axis Sharing

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(figsize=(5,5))
ax.set_title('Title', fontsize=20)
ax.set_xlabel('X Label',fontsize=15)
ax.set_ylabel('Y Label',fontsize=15)
```

### < 1. fig.tight_layout >

```python
title_list = ['Ax' + str(i) for i in range(4) ]
xlable_list = ['X label' + str(i) for i in range(4)]
ylable_list = ['Y label' + str(i) for i in range(4)]

print(title_list)
print(xlabel_list)
print(ylable_list)

fig, axes = plt.subplots(2, 2, figsize=(10,10))

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx],fontsize=20)
    ax.set_xlabel(xlabel_list[ax_idx],fontsize=15)
    ax.set_ylabel(ylabel_list[ax_idx],fontsize=15)
    
fig.tight_layout() # fig object의 layout을 tight하게 잡아준다.
# 간격을 좀 더 정확하게 설정하고 싶으면 괄호 안에 pad 값을 주면 된다. ex) fig.tight_layout(pad=3)
```

### < 2. fig.subplots_adjust > 

```python
# fig.sibplots_adgust( more customized layout )

fig, ax = plt.subplots(figsize = (5,5))
xaxis = ax.get_xaxis()
yaxis = ax.get_yaxis()

print(xaxis)
print(yaxis)
```

```python
# fig.sibplots_adgust( more customized layout )

fig, ax = plt.subplots(figsize = (5,5))
ax.get_xaxis().set_visible(False) # 종횡비와 축 없애기 set_visivle(False)
ax.get_yaxis().set_visible(False)
```

```python
fig, axes = plt.subplots(2,2, figsize=(2,2))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom = 0.01, top = 0.99)
```

```python
fig, axes = plt.subplots(2,2, figsize=(2,2))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom = 0.01, top = 0.99, left=0.01, right=0.99)
```

```python
fig, axes = plt.subplots(2,2, figsize=(2,2))

for ax_idx, ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
fig.subplots_adjust(bottom = 0.01, top = 0.99, left=0.01, right=0.99, hspace = 0.05, wspace = 0.05)
```

```python
import numpy as np

fig = plt.figure(figsize=(12,13))

axes_g1 = np.empty(shape=(0,))
main = plt.subplot2grid((5,4),(0,0),2,2,fig=fig)

axes_g1 = np.append(axes_g1, main)

for r_idx in range(2,2+3):
    for c_idx in range(2):
        ax = plt.subplot2grid((5,4),(r_idx,c_idx),fig=fig)
        axes_g1 = np.append(axes_g1, ax)
        
axes_g1 = axes_g1.reshape(1,-1)
print(axes_g1)

axes_g2 = np.empty(shape=(0,))
main = plt.subplot2grid((5,4),(0,2),2,2,fig=fig)

axes_g2 = np.append(axes_g2, main)

for r_idx in range(2,2+3):
    for c_idx in range(2,4):
        ax = plt.subplot2grid((5,4),(r_idx,c_idx),fig=fig)
        
        axes_g2 = np.append(axes_g2, ax)
        
axes_g2.reshape(1,-1)
print(axes_g2)

axes = np.vstack((axes_g1, axes_g2))
print(axes.shape)
```

#### reshpae 함수 

np.reshape(변경할 배열, 차원) 또는 배열.reshape(차원)으로 사용할 수 있으며, 현재의 배열의 차원(1차원,2차원,3차원)을 변경하여 <br>
행렬을 반환하거나 하는 경우에 많이 이용된다. 

```python
import numpy as np
a = [1,2,3,4,5,6,7,8]
b = np.reshape(a,(2,4))
c = np.reshape(a,(4,2))
d = np.reshape(a,(1,-1))
print(b)
print('\n')
print(c)
print('\n')
print(d)
```

reshape를 활용하는 경우를 보다보면 입력인수로 -1이 들어간 경우가 종종 있다.<br>
