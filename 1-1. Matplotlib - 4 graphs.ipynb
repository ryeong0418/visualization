```python
title_list=['Training Image', '2 scales', '4 scales', '5 scales', '6 scales', '8 scales']

fig,axes = plt.subplots(2,3,figsize=(20,10))

print(axes)
print(axes.flat)

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx], fontsize=20)
```

```python
# plt.subplot --> ndarray.flat and enumerate

test_np = np.array([[1,2],[3,4]])
print(test_np)
print('='*50)
print(test_np[0])
print(test_np[1])
```

```python
idx = 0
test_list=[10,11,12,13]

for val in test_list:
    print(idx,val)
    idx+=1
    
print('='*50)

for idx, val in enumerate(test_list):
    print(idx,val)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4,4), facecolor='linen')

for ax in axes.flat:
    print(ax)
    
print('='*50)

for ax_idx, ax in enumerate(axes.flat):
    print(ax_idx, ax)
```

##### < 3. subplot2grid >

```python
fig = plt.figure(figsize=(4,4), facecolor='linen')

ax1 = plt.subplot2grid((1,1),(0,0),fig=fig)
```

```python
fig = plt.figure(figsize=(7,7),facecolor='linen')

ax1 = plt.subplot2grid((3,3), (0,0), colspan=2, fig=fig)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, fig=fig)
ax3 = plt.subplot2grid((3,3), (2,0), colspan=2, fig=fig)
ax4 = plt.subplot2grid((3,3), (0,2), rowspan=3, fig=fig)
```

##### < 5. add_axes > 

```python
fig = plt.figure(figsize=(4,4),facecolor = 'linen')

ax = fig.add_axes([0.1,0.1,0.8,0.8]) # add_axes(left, bottom, width, height)
```

```python
left,bottom = 0.1, 0.1
spacing = 0.05
width1, height1 = 0.5, 0.5

width2 = 1-(2*left + width1 + spacing)
height2 = 1-(2*bottom + height1 + spacing)

rect1 = [left, bottom, width1, height1]
rect2 = [left, bottom+height1+spacing, 1-2*left, height2]
rect3 = [left+width1+spacing, bottom, width2, height1]


fig = plt.figure(figsize=(5,5), facecolor='linen')

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```

```python
left, bottom = 0.1, 0.1
spacing = 0.005

width1, height1 = 0.65, 0.65
width2 = 1 - (2*left + width1 + spacing)
height2 = 1 - (2*bottom + height1 + spacing)

rect1 = [left, bottom, width1, height1]
rect2 = [left, bottom+height1+spacing, width1, height2]
rect3 = [left+width1+spacing, bottom, width2, height1]

fig = plt.figure(figsize=(5,5))
ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```
